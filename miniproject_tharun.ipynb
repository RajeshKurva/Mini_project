```python
# Importing  necessary libraries
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd

# Importing dataset from kaggle.com
data_set = pd.read_csv('/content/water_potability.csv')

# Extracting independent variables (all columns except the last one)
x = data_set.iloc[:, :-1].values

# Extracting dependent variable (the last column, assuming it's the target)
y = data_set.iloc[:, -1].values

# Handling missing data (Replacing missing data with the mean value)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fitting imputer object to the independent variables x.
imputer = imputer.fit(x[:, 1:3])

# Replacing missing data with the calculated mean value
x[:, 1:3] = imputer.transform(x[:, 1:3])

# Encoding categorical data (if any)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

# Assuming first column has categorical data
label_encoder_x = LabelEncoder()
x[:, 0] = label_encoder_x.fit_transform(x[:, 0])

# Encoding dummy variables using OneHotEncoder
column_transformer = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough'
)
x = column_transformer.fit_transform(x)

# Encoding target variable (y), if it's categorical
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)

# Splitting the dataset into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)


scaler = StandardScaler(with_mean=False)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
# Checking the shapes of the training and test sets
print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
# Viewing the first 5 rows of x_train, x_test, y_train, and y_test
print("First 5 rows of x_train:\n", x_train[:5])
print("First 5 rows of x_test:\n", x_test[:5])
print("First 5 rows of y_train:\n", y_train[:5])
print("First 5 rows of y_test:\n", y_test[:5])
# Convert x_train and x_test to pandas DataFrame for a better view
x_train_df = pd.DataFrame(x_train)
x_test_df = pd.DataFrame(x_test)

print("First 5 rows of x_train DataFrame:\n", x_train_df.head())
print("First 5 rows of x_test DataFrame:\n", x_test_df.head())






```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

# Importing dataset from kaggle.com
data_set = pd.read_csv('/content/water_potability.csv')

# Extracting independent variables (all columns except the last one)
x = data_set.iloc[:, :-1].values

# Extracting dependent variable (the last column, assuming it's the target)
y = data_set.iloc[:, -1].values

# Handling missing data (Replacing missing data with the mean value)
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x = imputer.fit_transform(x)

# Encoding categorical variables
# Assuming the first column is categorical (adjust if necessary)
label_encoder_x = LabelEncoder()
x[:, 0] = label_encoder_x.fit_transform(x[:, 0])

# Applying OneHotEncoder to the categorical features
column_transformer = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough'
)
x = column_transformer.fit_transform(x)

# Encoding target variable (y), if it's categorical
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)

# Splitting the dataset into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Standardizing features
scaler = StandardScaler(with_mean=False)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Convert to DataFrame for saving
x_train_df = pd.DataFrame(x_train)
x_test_df = pd.DataFrame(x_test)
y_train_df = pd.DataFrame(y_train, columns=['Target'])
y_test_df = pd.DataFrame(y_test, columns=['Target'])

# Add identifier columns
x_train_df['Data_Type'] = 'Train'
x_test_df['Data_Type'] = 'Test'

# Concatenate features and target dataframes
train_df = pd.concat([x_train_df, y_train_df.reset_index(drop=True)], axis=1)
test_df = pd.concat([x_test_df, y_test_df.reset_index(drop=True)], axis=1)

# Combine train and test DataFrames
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Save to a single CSV file
combined_df.to_csv('/content/combined_preprocessed_data.csv', index=False)

print("Combined CSV file for preprocessed data has been saved.")
```
